### (1) Erosion, Dilation, Opening and Closing.	

(a) Load the RGB image clock2.jpg. Convert it to grayscale and then convert it into a binary image using a threshold of 0.5.

(b) Use an SE as [1 1 1; 1 1 1; 1 1 1] to obtain the binary morphological erosion, dilation, opening and closing of the binary image. Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

(c) Use an SE as [1 0 0; 0 1 0; 0 0 1] to obtain the binary morphological erosion, dilation, opening and closing of the binary image. Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

(d) Load the RGB image zebras.jpg. Convert it to grayscale. 

(e) Use a 3x3 SE to obtain the grayscale morphological erosion, dilation, opening and closing of grayscale image. Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

(f) Use a 7x7 SE to obtain the grayscale morphological erosion, dilation, opening and closing of grayscale image. Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

Hint:``skimage.morphology.binary_erosion``, ``skimage.morphology.binary_dilation``, ``skimage.morphology.binary_opening``, ``skimage.morphology.binary_closing``, ``skimage.morphology.erosion``, ``skimage.morphology.dilation``, ``skimage.morphology.opening`` and ``skimage.morphology.closing``

````python
from skimage import io, color
import numpy as np
from skimage.morphology import binary_erosion, binary_dilation, binary_opening, binary_closing
from skimage.morphology import erosion, dilation, opening, closing, square
import matplotlib.pyplot as plt

# Load, convert to grayscale, then convert to binary
img1_1 = io.imread('./images/clock2.jpg')
img1_1 = color.rgb2gray(img1_1)
img1_binary = np.where(img1_1>0.5, 1, 0)

# Erosion, dilation, opening, closing
se_1 = np.array([[1,1,1],[1,1,1],[1,1,1]])
img_1_er = binary_erosion(img1_binary, se_1)
img_1_di = binary_dilation(img1_binary, se_1)
img_1_op = binary_opening(img1_binary, se_1)
img_1_cl = binary_closing(img1_binary, se_1)

plt.figure(figsize=[12, 10])
plt.subplot(3, 2, 1)
plt.imshow(img1_binary, cmap='gray')
plt.title('Origin')
plt.subplot(3, 2, 3)
plt.imshow(img_1_er, cmap='gray')
plt.title('Erosion (3*3 suqare SE)')
plt.subplot(3, 2, 4)
plt.imshow(img_1_di, cmap='gray')
plt.title('Dilation (3*3 suqare SE)')
plt.subplot(3, 2, 5)
plt.imshow(img_1_op, cmap='gray')
plt.title('Opening (3*3 suqare SE)')
plt.subplot(3, 2, 6)
plt.imshow(img_1_cl, cmap='gray')
plt.title('Closing (3*3 suqare SE)')
plt.tight_layout()
plt.show()

se_2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
img_2_er = binary_erosion(img1_binary, se_2)
img_2_di = binary_dilation(img1_binary, se_2)
img_2_op = binary_opening(img1_binary, se_2)
img_2_cl = binary_closing(img1_binary, se_2)

plt.figure(figsize=[12, 10])
plt.subplot(3, 2, 1)
plt.imshow(img1_binary, cmap='gray')
plt.title('Origin')
plt.subplot(3, 2, 3)
plt.imshow(img_2_er, cmap='gray')
plt.title('Erosion (3*3 diag SE)')
plt.subplot(3, 2, 4)
plt.imshow(img_2_di, cmap='gray')
plt.title('Dilation (3*3 diag SE)')
plt.subplot(3, 2, 5)
plt.imshow(img_2_op, cmap='gray')
plt.title('Opening (3*3 diag SE)')
plt.subplot(3, 2, 6)
plt.imshow(img_2_cl, cmap='gray')
plt.title('Closing (3*3 diag SE)')
plt.tight_layout()
plt.show()

# Load, convert to grayscale
img1_2 = io.imread('./images/zebras.jpg')
img1_2 = color.rgb2gray(img1_2)

img_3_er = erosion(img1_2, square(3))
img_3_di = dilation(img1_2, square(3))
img_3_op = opening(img1_2, square(3))
img_3_cl = closing(img1_2, square(3))

plt.figure(figsize=[12, 10])
plt.subplot(3, 2, 1)
plt.imshow(img1_2, cmap='gray')
plt.title('Origin')
plt.subplot(3, 2, 3)
plt.imshow(img_3_er, cmap='gray')
plt.title('Erosion (3*3 square SE)')
plt.subplot(3, 2, 4)
plt.imshow(img_3_di, cmap='gray')
plt.title('Dilation (3*3 square SE)')
plt.subplot(3, 2, 5)
plt.imshow(img_3_op, cmap='gray')
plt.title('Opening (3*3 square SE)')
plt.subplot(3, 2, 6)
plt.imshow(img_3_cl, cmap='gray')
plt.title('Closing (3*3 square SE)')
plt.tight_layout()
plt.show()

img_4_er = erosion(img1_2, square(7))
img_4_di = dilation(img1_2, square(7))
img_4_op = opening(img1_2, square(7))
img_4_cl = closing(img1_2, square(7))

plt.figure(figsize=[12, 10])
plt.subplot(3, 2, 1)
plt.imshow(img1_2, cmap='gray')
plt.title('Origin')
plt.subplot(3, 2, 3)
plt.imshow(img_4_er, cmap='gray')
plt.title('Erosion (7*7 square SE)')
plt.subplot(3, 2, 4)
plt.imshow(img_4_di, cmap='gray')
plt.title('Dilation (7*7 square SE)')
plt.subplot(3, 2, 5)
plt.imshow(img_4_op, cmap='gray')
plt.title('Opening (7*7 square SE)')
plt.subplot(3, 2, 6)
plt.imshow(img_4_cl, cmap='gray')
plt.title('Closing (7*7 square SE)')
plt.tight_layout()
plt.show()
````

### (2) Boundary Extraction and Hole Filling

(a) Load the image boundary.tif. Convert it into a binary image.

(b) Use a 5x5 SE of 1s to obtain the binary morphological erosion of the binary image.

(c) Obtian the boundary of the binary image via performing the difference between the binary image and its erosion image.

(d) Load the image holefiling.tif. Convert it into a binary image.

(e) Fill the holes in binary objects. (Hine:``scipy.ndimage.binary_fill_holes``)

(f) Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

````python
import cv2
from scipy.ndimage import binary_fill_holes

# Load and convert
img2_1 = cv2.imread('./images/boundary.tif')
img2_1 = cv2.cvtColor(img2_1, cv2.COLOR_BGR2RGB)
img2_1 = color.rgb2gray(img2_1)
img2_1_binary = np.where(img2_1>0.5, 1, 0)

# Erosion
img2_er = binary_erosion(img2_1_binary, square(5))
# Get boundary
img2_boun = img2_1_binary - img2_er

plt.figure(figsize=[12, 8])
plt.subplot(1, 3, 1)
plt.imshow(img2_1_binary, cmap='gray')
plt.title('Origin')
plt.subplot(1, 3, 2)
plt.imshow(img2_er, cmap='gray')
plt.title('Erosion')
plt.subplot(1, 3, 3)
plt.imshow(img2_boun, cmap='gray')
plt.title('Boundary')
plt.tight_layout()
plt.show()

# Load and convert
img2_2 = cv2.imread('./images/holefiling.tif')
img2_2 = cv2.cvtColor(img2_2, cv2.COLOR_BGR2RGB)
img2_2 = color.rgb2gray(img2_2)
img2_2_binary = np.where(img2_2>0.5, 1, 0)

# Fill the holes
img2_fill = binary_fill_holes(img2_2_binary)

plt.figure(figsize=[12, 8])
plt.subplot(1, 2, 1)
plt.imshow(img2_2_binary, cmap='gray')
plt.title('Origin')
plt.subplot(1, 2, 2)
plt.imshow(img2_fill, cmap='gray')
plt.title('Holes filled')
plt.tight_layout()
plt.show()
````

### (3) Thining and Skeleton
(a) Load the image CT.tif. Convert it into a binary image.

(b) Perform morphological thinning of a binary image. (Hint:``skimage.morphology.thin``, set the parameter of max_iter to 5, 50, and none, respectively.</font>)

(c) Compute the convex hull image of a binary image. (Hint:``skimage.morphology.convex_hull_image``)

(d) Compute the skeleton of a binary image. (Hint:``skimage.morphology.skeletonize``)

(e) Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

````python
from skimage.morphology import thin, convex_hull_image, skeletonize

# Load and convert
img3 = cv2.imread('./images/CT.tif')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img3 = color.rgb2gray(img3)
img3_binary = np.where(img3>0.5, 1, 0)

# Thin
img3_thin_1 = thin(img3_binary, max_num_iter=5)
img3_thin_2 = thin(img3_binary, max_num_iter=50)
img3_thin_3 = thin(img3_binary, max_num_iter=None)

# Convex hull
img3_convex = convex_hull_image(img3_binary)

# Sketelon
img3_sketelon = skeletonize(img3_binary)

plt.figure(figsize=[12, 8])
plt.subplot(2, 3, 1)
plt.imshow(img3_binary, cmap='gray')
plt.title('Origin')
plt.subplot(2, 3, 2)
plt.imshow(img3_thin_1, cmap='gray')
plt.title('Thinning (iteration = 5)')
plt.subplot(2, 3, 3)
plt.imshow(img3_thin_2, cmap='gray')
plt.title('Thinning (iteration = 50)')
plt.subplot(2, 3, 4)
plt.imshow(img3_thin_3, cmap='gray')
plt.title('Thinning (iterate until converge)')
plt.subplot(2, 3, 5)
plt.imshow(img3_convex, cmap='gray')
plt.title('Convex hull image')
plt.subplot(2, 3, 6)
plt.imshow(img3_sketelon, cmap='gray')
plt.title('Skeleton')
plt.tight_layout()
plt.show()
````

### (4) Top-hat (white top-hat) and Bottom-hat (black top-hat) Transformation

(a) Load the grayscale image rice.tif. Convert it into a binary image with a threshold of 0.5.

(b) Perform top-hat transformation to the grayscale image. (Hint:``skimage.morphology.white_tophat``, set the parameter of footprint to ``skimage.morphology.square(81)``)

(c) Obtain thresholded top-hat image and convert it into a binary image.

(d) Perform bottom-hat transformation to the grayscale image. (Hint:``skimage.morphology.black_tophat``, set the parameter of footprint to ``skimage.morphology.square(81)``)

(e) Obtain thresholded bottom-hat image and convert it into a binary image.

(f) Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.


````python
from skimage import img_as_float
from skimage.morphology import white_tophat, black_tophat

# Load and convert
img4 = io.imread('./images/rice.tif')
img4 = img_as_float(img4)
img4_binary = np.where(img4>0.5, 1, 0)

# Top-hat
img4_top = white_tophat(img4, footprint=square(81))
# Threshold
img4_top_binary = np.where(img4_top>0.2, 1, 0)

# Bottom-hat
img4_bottom = black_tophat(img4, footprint=square(81))
# Threshold
img4_bottom_binary = np.where(img4_bottom>0.2, 1, 0)

plt.figure(figsize=[12, 12])
plt.subplot(3, 2, 1)
plt.imshow(img4, cmap='gray')
plt.title('Original image')
plt.subplot(3, 2, 2)
plt.imshow(img4_binary, cmap='gray')
plt.title('Thresholded original image')
plt.subplot(3, 2, 3)
plt.imshow(img4_top, cmap='gray')
plt.title('Top-hat image')
plt.subplot(3, 2, 4)
plt.imshow(img4_top_binary, cmap='gray')
plt.title('Thresholded top-hat image')
plt.subplot(3, 2, 5)
plt.imshow(img4_bottom, cmap='gray')
plt.title('Bottom-hat image')
plt.subplot(3, 2, 6)
plt.imshow(img4_bottom_binary, cmap='gray')
plt.title('Thresholded bottom-hat image')
plt.tight_layout()
plt.show()
````

### (7) Hough transform
(a) Load the image triangle_circle.png. Convert it to grayscale.

(b) Perform a straight line Hough transform to the grayscale image. (Hint:``skimage.transform.hough_line``)

(c) Obtain the peaks in a straight line Hough transform. (Hint:``skimage.transform.hough_line_peaks``)

(d) Highlight the detected lines by red color. (Hint:``matplotlib.pyplot.axline``)

(e) Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.



````python
from skimage.transform import hough_line, hough_line_peaks

img7 = cv2.imread('./images/triangle_circle.png')
img7 = cv2.cvtColor(img7, cv2.COLOR_BGR2GRAY)

# Straight line Hough transform
h, theta, rou = hough_line(img7)

plt.figure(figsize=[8, 8])
plt.subplot(2, 2, 1)
plt.imshow(img7, cmap='gray')
plt.title('Original image')
plt.subplot(2, 2, 2)
plt.imshow(np.log(1 + h), cmap='gray')
plt.xlabel('Angles (degrees)')
plt.ylabel('Distance (pixels)')
plt.title('Hough transform')
plt.subplot(2, 2, 3)
plt.imshow(img7, cmap='gray')
plt.title('Detected lines')

# Obtain the peaks and highlight the detected lines
for h_, theta_, rou_ in zip(*hough_line_peaks(h, theta, rou)):
    if np.sin(theta_) != 0:
        intercept = rou_ / np.sin(theta_)
        slope = -1 / np.tan(theta_)
        plt.axline((0, intercept), slope=slope, color='red', lw=3)
    else:
        x = rou_ / np.cos(theta_)
        plt.axline((x, 0), (x, 1), color='red', lw=3)

plt.tight_layout()
plt.show()
````

### (8) Thresholding-Based Segmentation 
(a) Load the image shade_text1.tif. 

(b) Perform automatic image thresholding by Otsu's method to get the threshold value. (Hint:``skimage.filters.threshold_otsu``)

(c) Transform the grayscale image into a binary image using the threshold value obtained by Otsu's method.

(d) Compute a moving average threshold. (Hint:``cv2.blur``)

(e) Obtain the result of local thresholding using moving averages.

(f) Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

````python
from skimage.filters import threshold_otsu

img8 = cv2.imread('./images/shade_text1.tif')
img8 = cv2.cvtColor(img8, cv2.COLOR_BGR2GRAY)

# Get the threshold value by Otsu's method
thresh = threshold_otsu(img8)

# The binary image using the threshold value
thresh_img = np.zeros(img8.shape)
thresh_img[img8 > thresh] = 1

# The moving average threshold
img8_blur = cv2.blur(img8, (3, 3))
thresh_img_blur = (img8 >= img8_blur*0.2) * 1.0

plt.figure(figsize=[8, 8])
plt.subplot(2, 2, 1)
plt.imshow(img8, cmap='gray')
plt.title('Original image')
plt.subplot(2, 2, 3)
plt.imshow(thresh_img, cmap='gray')
plt.title('Otsu\'s method threshold segmentation')
plt.subplot(2, 2, 4)
plt.imshow(thresh_img_blur, cmap='gray')
plt.title('The moving average threshold segmentation')
plt.tight_layout()
plt.show()
````

### (9) Segmentation Using Morphological Watersheds
(a) Load the coins image by skimage. (Hint:``skimage.data.coins()``)

(b) Find edges as the elevation map in an image using the Sobel filter. (Hint:``skimage.filters.sobel``)

(c) Define the markers as follow:
```python
markers = np.zeros_like(coins)
markers[coins < 30] = 1
markers[coins > 150] = 2
```
(d) Obtain the segmentation result by morphological Watersheds.  (Hint:``skimage.morphology.watershed``)

(e) Display them in the same figure with sub-figures. Add the corresponding title to the sub-figure.

````python
from skimage.segmentation import watershed

img9 = coins()

# Find edges as the elevation map
img9_elevation = sobel(img9)

# Define the markers
markers = np.zeros_like(img9)
markers[img9 < 30] = 1
markers[img9 > 150] = 2

# Segmentation
img9_seg = watershed(img9_elevation, markers)

plt.figure(figsize=[8, 8])
plt.subplot(2, 2, 1)
plt.imshow(img9, cmap='gray')
plt.title('Original image')
plt.subplot(2, 2, 2)
plt.imshow(img9_elevation, cmap='gray')
plt.title('Elevation map')
plt.subplot(2, 2, 3)
plt.imshow(markers, cmap='gray')
plt.title('Markers')
plt.subplot(2, 2, 4)
plt.imshow(img9_seg, cmap='gray')
plt.title('Watershed segmentation')
plt.tight_layout()
plt.show()
````
